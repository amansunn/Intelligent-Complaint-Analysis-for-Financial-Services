Initial EDA – Understand the Data

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('../data/raw/complaints.csv')

# Basic shape
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])

# Display column names
print("\nColumns:")
print(df.columns.tolist())

# Data types and non-null counts
print("\nDataset info:")
df.info()

# Missing value summary
print("\nMissing values per column:")
print(df.isnull().sum())

# Sample few rows
df.sample(5)
```

Analyze distribution of complaints by Product

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count complaints by product
product_counts = df['Product'].value_counts()

# Display counts
print("Complaints per product:")
print(product_counts)

# Plot distribution
plt.figure(figsize=(12, 6))
sns.barplot(x=product_counts.index, y=product_counts.values, palette='viridis')
plt.title('Number of Complaints by Product')
plt.xlabel('Product')
plt.ylabel('Number of Complaints')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Analyze Complaint Narrative Lengths

```python
# Ensure "Consumer complaint narrative" column is string type
df['Consumer complaint narrative'] = df['Consumer complaint narrative'].astype(str)

# Calculate word count
df['narrative_word_count'] = df['Consumer complaint narrative'].apply(lambda x: len(x.split()))

# Basic stats
print("Narrative word count summary:")
print(df['narrative_word_count'].describe())

# Histogram of word counts
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
df['narrative_word_count'].hist(bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Consumer Complaint Narrative Lengths')
plt.xlabel('Word Count')
plt.ylabel('Number of Complaints')
plt.grid(False)
plt.tight_layout()
plt.show()
```

 Identify the Number of Complaints With and Without Narratives

```python
# Count missing (NaN or empty) narratives
df['has_narrative'] = df['Consumer complaint narrative'].apply(lambda x: isinstance(x, str) and x.strip() != '')

# Summary
with_narrative = df['has_narrative'].sum()
without_narrative = len(df) - with_narrative

print(f"Complaints WITH narratives: {with_narrative}")
print(f"Complaints WITHOUT narratives: {without_narrative}")
print(f"Percentage with narratives: {with_narrative / len(df) * 100:.2f}%")
```

Filter Dataset & Clean Complaint Narratives

```python
import re

# 1. Filter for specific products
target_products = [
    'Credit card', 'Personal loan',
    'Buy Now, Pay Later (BNPL)',
    'Savings account', 'Money transfers'
]

filtered_df = df[
    df['Product'].isin(target_products) &
    df['has_narrative']  # already created earlier
].copy()

print(f"Remaining records after filtering: {len(filtered_df)}")

# 2. Clean text function
def clean_text(text):
    # Lowercase
    text = text.lower()
    # Remove boilerplate phrases (customize this as needed)
    text = re.sub(r"i am writing to file a complaint.*?\.", '', text)
    text = re.sub(r"^dear [a-z\s,]+", '', text)
    # Remove non-alphabetic characters except basic punctuation
    text = re.sub(r"[^a-z0-9\s.,!?$%/-]", '', text)
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# 3. Apply cleaning
filtered_df['cleaned_narrative'] = filtered_df['Consumer complaint narrative'].apply(clean_text)

# Final check
print(filtered_df[['Product', 'cleaned_narrative']].sample(3))
```

```python
# Save cleaned and filtered data
filtered_df.to_csv('../data/filtered_complaints.csv', index=False)
print("Saved to data/filtered_complaints.csv")
```
